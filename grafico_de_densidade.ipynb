```python
dadostrees = pd.read_csv('trees.csv')
dadostrees.head()
```

```python
dadostrees.rename(columns={'Girth': 'Circunferência', 'Heigth': 'Altura'}, inplace=True)
dadostrees.head()
```

```python
# a fim de comparação, vou plotar um histograma

plt.hist(dadostrees.iloc[:,1], bins=6)
plt.title('Árvores')
```

```python
# vou plotar agora com o seaborn, e vou usar a linha de densidade
# no histograma, usando o 'kde'

sns.histplot(dadostrees.iloc[:,1], bins=6, kde=True, color='blue').set(title='Árvores')
```

```python
# agora vou plotar apenas a linha de densidade
sns.kdeplot(dadostrees.iloc[:,1]).set(title='Árvores')
```
