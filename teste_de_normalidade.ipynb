```python
from scipy import stats
from scipy.stats import norm, skewnorm
import matplotlib.pyplot as plt
```

```python
# vou criar dados com uma distribuição normal e outra, depois
# vou a distribuição assimétrica, para comparação
# irei aplicar 3 formas de testar a normalidade dos dados

dados_normais = norm.rvs(size = 1000)
```

```python
# a primeira forma é gerando um histograma

plt.hist(dados_normais, bins = 20 )
plt.title('Dados ')
```

```python
# a segunda forma é gerando um qqplot
fig, ax = plt.subplots()

# fig significa que to criando uma figura vazia e ax que tô especificando um eixo dela
stats.probplot(dados_normais, fit = True, plot = ax)
plt.show
```

```python
# a terceira forma é executando o teste de shapiro, se o valor de 'p' for maior que
# 0.05 significa que a hipotese nula é verdadeira, logo os dados são normais

stats.shapiro(dados_normais)
```
