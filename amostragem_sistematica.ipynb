```python
import pandas as pd
import numpy as np
from math import ceil
# a função ceil é pra pegar o maior valor da divisão de um número real
```

```python
populacao = 150
amostra = 15
# vamos calcular o espaço de seleção da amostra
intervalo = ceil( populacao / amostra)
intervalo
```

```python
# definindo um valor aleatório para começar a seleção, de 1 até o valor do intervalo
inicio = np.random.randint( low = 1, high = intervalo, size=1)
inicio
```

```python
# vou ficar um for pra somar os valores no espaço do intervalo que calculei
# começando pelo valor aleatório que gerei

acumulador = inicio[0]
sorteados = []
for i in range(amostra):
  sorteados.append(acumulador)
  acumulador += intervalo
print(sorteados)
```
