```python
# filtro usandp uma condição de valor

dados.loc[dados['purpose'] == "radio/tv"]
```

```python
# utilizando outra condição para filtrar

dados.loc[dados['credit_amount'] > 18000]
```

```python
# atribuindo resultado a uma variável, criando outro dataframe

credito = dados.loc[dados['credit_amount']> 18000]
print(credito)
```

```python
# aplicando o mesmo filtro mas selecionando apenas algumas colunas

credito2 = dados[['class', 'job']].loc[dados['credit_amount']>1800]
print(credito2)
```

```python
# séries podem ser criadas a partir de listas, arrays do numpy ou colunas de dataframe

s1 = pd.Series([7,5,9,9,8,8,5,6,2,5])
print(s1)
```

```python
# série a partir de um array do numpy

import numpy as np
array = np.array([1,5,6,3,2,7,8,9])
s2 = pd.Series(array)
print(s2)
```

```python
# série a partir de um dataframe

s3 = dados['purpose']
print(s3)
type(s3)
```

```python
# observa-se a diferença, com um "[]" a mais se torna um dataframe e não uma série

dataf = dados[['purpose']]
type(dataf)
```

```python
# renomeando colunas
dados.rename(columns={'purpose': 'propósito'}, inplace=True)
dados.head()
```

```python
# excluindo coluna (axis=0 é o eixo de linhas, axis=1 é o eixo das colunas)
dados.drop('checking_status', axis=1, inplace=True)
dados.head()
```

```python
# checando a existência de dados nulos

dados.isnull()
```

```python
# verificando a quantidade de dados nulos
dados.isnull().sum()
```

```python
# preenchendo dados faltantes com "0"

dados['credit_amount'].fillna(0, inplace=True)
```

```python
# selecionando para exibir determinado intervalo de linhas e colunas

dados.iloc[0:3,0:5]
```

```python
# indicando as linhas como lista
dados.iloc[[1,2,3,4,5,6], 0:5]
```
