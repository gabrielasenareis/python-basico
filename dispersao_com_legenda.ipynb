```python
base = pd.read_csv('co2.csv')
base.head()

# essa base de dados é de um estudo de plantas que recebem um
# tratamendo refrigerado e não refrigerado, e sua concentração de
# co2 e absorção de co2 a depender do tipo de tratamento
```

```python
base.rename(columns={'Plant': 'Planta', 'Type': 'Tipo', 'Treatment': 'Tratamento', 'conc': 'Concentração', 'uptake': 'Absorção' }, inplace=True)
base.head()
```

```python
base.loc[base['Tratamento'] == 'nonchilled', 'Tratamento'] = "Não refrigerada"
base.loc[base['Tratamento'] == 'chilled', 'Tratamento'] = "Refrigerada"
base.head()
```

```python
# vou fazer um gráfico de dispersão dos valores de concentração e absorção de co2
# em relação ao tipo de tratamento, colocando a legenda
x = base.Concentração
y = base.Absorção
tratamentos = list(set(base.Tratamento))
tratamentos
```

```python
 # criando o gráfico, vou criar uma função pra percorrer os
 # dois valores do tratamento, e imprimir o gráfico com
 # cada tratamento

for tratamento in tratamentos:
    indice = base.Tratamento == tratamento
    plt.scatter(x[indice], y[indice], label = tratamento)
plt.xlabel('Concentração')
plt.ylabel('Absorção')
# posicionamento da legenda
plt.legend(loc = 'lower right')
```
