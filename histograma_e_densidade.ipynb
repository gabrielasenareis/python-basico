```python
base2 = pd.read_csv('chicken.csv')
base2.head()
```

```python
base2.rename(columns={'weight': 'Peso', 'feed': 'Ração'}, inplace=True)
base2.head()
```

```python
sns.histplot(x=base2.Peso, y=base2.Ração, color = 'green', kde= True, bins = 10)
```

```python
# também posso fazer esses gráficos separados e junta-los lado a lado para fazer
# uma comparação melhor de valores

agrupado = base2.groupby(['Ração']).sum()
agrupado
```
