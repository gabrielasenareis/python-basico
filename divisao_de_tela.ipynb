```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
base = pd.read_csv('trees.csv')
base.rename(columns={'Girth': 'Circunferência', 'Height': 'Altura'}, inplace=True)
base.head()
```

```python
# vou criar os gráficos lado a lado, para facilitar a comparação deles

plt.figure(1)
#posição do primeiro grafico
plt.subplot(2,2,1)
plt.scatter(base.Circunferência, base.Volume)
plt.xlabel('Circunferência')
plt.ylabel('Volume')
# posição do segundo gráfico
plt.subplot(2,2,2)
plt.scatter(base.Circunferência, base.Altura)
plt.xlabel('Circunferência')
plt.ylabel('Altura')
# posição do terceiro gráfico
plt.subplot(2,2,3)
plt.scatter(base.Altura, base.Volume)
plt.xlabel('Altura')
plt.ylabel('Volume')
# posição do quarto gráfico
plt.subplot(2,2,4)
plt.hist(base.Volume)
plt.xlabel('Volume')
plt.ylabel('Frequência')
# as informações do eixo x e y estão sendo sobrepostas pelos gráficos
# vou aumentar o espaçamento entre eles usando a função plt.subplots_adjust()
```

```python
plt.figure(1)
#posição do primeiro grafico
plt.subplot(2,2,1)
plt.scatter(base.Circunferência, base.Volume)
plt.xlabel('Circunferência')
plt.ylabel('Volume')

# função para espaçamento
plt.subplots_adjust(hspace=0.5)
plt.subplots_adjust(wspace=0.5)
# posição do segundo gráfico
plt.subplot(2,2,2)
plt.scatter(base.Circunferência, base.Altura)
plt.xlabel('Circunferência')
plt.ylabel('Altura')
# posição do terceiro gráfico
plt.subplot(2,2,3)
plt.scatter(base.Altura, base.Volume)
plt.xlabel('Altura')
plt.ylabel('Volume')
# posição do quarto gráfico
plt.subplot(2,2,4)
plt.hist(base.Volume)
plt.xlabel('Volume')
plt.ylabel('Frequência')
```
