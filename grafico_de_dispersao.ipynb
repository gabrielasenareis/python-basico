```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dadostrees = pd.read_csv('trees.csv')
dadostrees.head()
```

```python
dadostrees.rename(columns={'Girth': 'Circunferência', 'Height': 'Altura'}, inplace=True)
dadostrees.head()
```

```python
# vou plotar um gráfico de disperção, que é um gráfico de comparação
# entre duas variáveis
plt.scatter(x=dadostrees.Circunferência, y=dadostrees.Volume, color='red', marker='*')
plt.xlabel('Circunferência')
plt.ylabel('Volume')
plt.title('Árvores')
# observamos que quando aumenta o volume, aumenta também a circunferência
```

```python
# agora um gráfico sem definir o 'marker', ele vai virar um gráfico de linha
# porque conecta os pontos

plt.plot(dadostrees.Circunferência, dadostrees.Volume, color='red')
plt.title('Árvores')
plt.xlabel('Circunferência')
plt.ylabel('Volume')
```

```python
# agora um gráfico de dispersão com afastamento dos dados, utilizando 'jitter'
# o 'fit reg' é uma linha de tendência, vou gerar outro gráfico com ele recebendo
# 'true' em seguida

sns.regplot(x=dadostrees.Circunferência, y=dadostrees.Volume, data = dadostrees, x_jitter = 0.3, fit_reg = False).set(title='Árvores')
```

```python
# agora utilizando a linha de frequência
sns.regplot(x=dadostrees.Circunferência, y=dadostrees.Volume, data = dadostrees, x_jitter = 0.3, fit_reg = True).set(title='Árvores')
```

```python
plt.scatter(x=[dadostrees.Circunferência == 8.3], y=[dadostrees.Circunferência == 8.3], color='red', marker='*')
plt.xlabel('Circunferência')
plt.ylabel('Volume')
```
