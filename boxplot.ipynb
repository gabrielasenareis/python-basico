```python
base = pd.read_csv('trees.csv')
base.rename(columns={'Girth': 'Circunferência', 'Height': 'Altura'}, inplace=True)
base.head()
```

```python
# vou gerar um boxplot da coluna de 'volume'
# showfliers é para mostrar valores outliers e o patch_artist
# para preencher a cor

plt.boxplot(base.Volume, vert = False, showfliers = True, patch_artist = True)
plt.title('Árvores')
plt.xlabel('Volume')
```

```python
# vou gerar 3 boxplots, cada um mostrando informações diferentes
plt.boxplot(base.Volume, vert = False, patch_artist = True)
plt.boxplot(base.Altura, vert = False, patch_artist = True)
plt.boxplot(base.Circunferência, vert = False, patch_artist = True)
plt.title('Árvores')
plt.xlabel('Dados')
```
