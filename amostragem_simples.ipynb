```python
# criando a semente randômica pra ter sempre o mesmo resultado aleatório
np.random.seed(2345)
# como existem 150 elementos sendo avaliados, vou fazer uma seleção aléatória de
# 70% e 30% desses itens (não necessariamente vai ser certinho 70 e 30%)
# vou gerar uma matriz aleatória com 70% de '0' e 30% de '1'
# o replace tem que receber true, senão não conseguiria gerar '0' e '1' repetidos

amostra = np.random.choice(a = [0,1], size = 150, replace=True, p = [0.7, 0.3])
amostra
```

```python
# vamos verificar se realmente o tamanho dos '0' é de 70%
len(amostra[amostra == 0])
```

```python
numero_total = 150
parte = 101
porcentagem = (parte / numero_total) * 100
porcentagem
# está próximo de 70% então funcionou normal
```

```python
# agora eu vou selecionar os itens no dataframe, de acordo com a posição aleatória
# da matriz de '0' e '1' que criei

amostragem70 = base.loc[amostra == 0]
amostragem70
```
