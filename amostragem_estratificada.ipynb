```python
import pandas as pd
from sklearn.model_selection import train_test_split
# o train test split aqui é pra fazer a divisão da base de dados
```

```python
base = pd.read_csv('iris.csv')
base.head()
```

```python
# vou listar quantas classes existem, e fazer a separação para
# amostra estratificada por cada classe
base['class'].value_counts()
```

```python
# vou usar train_test_split pra dividir em partes iguais a classe, ela retorna
# 4 objetos, mas podemos usar só 2, como vai ser nesse caso, o 'X e y'
# o primeiro parâmetro é o dataframe que vai ser usado para a divisão
# o segundo parâmetro é o rótulo ou variável dependente, onde  para cada X vai ter o y
# o 'test size' é a porcentagem da amostra,  e o 'stratify' é em que argumento vai ser
# realizada a divisão
X,_,y,_ = train_test_split(base.iloc[:,0:5], base.iloc[:,4], test_size=0.5, stratify = base.iloc[:,4])

# vou verificar se realmente foi selecionados 25 valores
y.value_counts()
```

```python
# outras forma de verificar
filtro = X['class'] == 'Iris-setosa'
serie_filtrada = pd.Series(filtro)
serie_filtrada.sum()
```
